```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
```

```python
dados = pd.read_csv("/content/A142052189_28_143_208.csv", encoding="ISO-8859-1", skiprows=3, sep=";", skipfooter=12, thousands='.',decimal=',')
dados.head()
```

```python
#Formatando os valores flutuantes para duas casas decimais
pd.options.display.float_format = "{:.2f}".format
```

```python
dados.mean() #retorna valores apenas para colunas que tem todos os valores
```

```python
colunas_usaveis = dados.mean().index.tolist()
colunas_usaveis.insert(0,'Unidade da Federação')
colunas_usaveis
```

```python
usaveis = dados[colunas_usaveis]
usaveis = usaveis.set_index('Unidade da Federação')
usaveis = usaveis.drop('Total', axis=1)

usaveis.head()
```

#Desafio 01: Reposicionar a legenda do gráfico em uma posição mais adequada

```python
ax = usaveis.T.plot(figsize=(13,9))
ax.legend(loc='best', bbox_to_anchor=(0.8, 0.5, 0.5, 0.5))
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("R$ {x:,.2f}"))

plt.title("SUS: Evolução dos gastos Hospitalares em R$")
plt.show()
```

#Desafio 02: Plotar o gráfico de linha com apenas 5 estados de sua preferência

```python
cinco_estados = usaveis.loc [['32 Espírito Santo', '33 Rio de Janeiro',
                              '35 São Paulo', '31 Minas Gerais', '53 Distrito Federal']]
```

```python
ax = cinco_estados.T.plot(figsize =(13, 9))
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("R$ {x:,.2f}"))

plt.title("SUS: Evolução dos gastos Hospitalares em R$")
plt.show()
```
