```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
```

```python
dados = pd.read_csv('/content/A123014189_28_143_208.csv', encoding='ISO-8859-1', skiprows=3, sep=';', skipfooter=12, thousands='.',decimal=',')
pd.options.display.float_format = "{:.2f}".format
dados
```

# Desafio 01: Escolher um título mais descritivo, que passe a mensagem adequada ao gráfico de barras.

```python
ax = dados.plot(x="Unidade da Federação", y="2008/Ago", kind="bar", figsize=(9,6))
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("R$ {x:,.2f}"))

plt.title("SUS: Gastos Hospitalares em R$")
plt.show()
```
